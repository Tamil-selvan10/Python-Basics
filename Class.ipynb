### Methods

```python
class Person:
    home='earth' # class object attribute
    def __init__(self,name,age):
        self.name=name #attributes
        self.age=age   #attributes
        
    def learn(self,subject): # methods
        print(f'I am {self.name} and my subject is {subject}')
```

```python
p1.name
```

```python
p1.age
```

```python
p1.learn('Economics')
```

### Inheritance



```python
class Person:
    
    def __init__(self,name):
        self.name=name
        
    def display(self):
        print(f'Name : {self.name}')
    
```

```python
p1=Person('Vijay')
```

```python
p1.name
```

```python
p1.display()
```

```python
class Student(Person):
    
    def __init__(self,name,grade):
        super().__init__(name)
        self.grade=grade
        
    def status(self): # new method
        print('Good')
    
    
    def display(self): # redefine method
        super().display()
        print(f'Grade : {self.grade}')
```

```python
s1=Student('Vijay','A+')
```

```python
dir(s1)
```

```python
s1.name
```

```python
s1.grade
```

```python
s1.display()
```

```python
s1.status()
```

### Polymorphism


```python
class Dog:
    
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        print(f'{self.name} says woof !') 
```

```python
class Cat:
    
    def __init__(self,name):
        self.name=name
        
    def speak(self):
        print(f'{self.name} says meou !')
```

```python
d1=Dog('max')
```

```python
c1=Cat('Tom')
```

```python
d1.speak()
```

```python
c1.speak()
```

### Abstract Classes

```python
class Document:
    
    def __init__(self,name):
        self.name=name
        
    def show(self):
        raise NotImplementedError('The subclass must implement abstract method')
```

```python
d1=Document('pdf')
```

```python
d1.name
```

```python
d1.show()
```

```python
class Pdf(Document):
    
    def show(self):
        return 'pdf content is here !'
```

```python
class Word(Document):
    
    def show(self):
        return 'Word content is here !'
```

```python
p1=Pdf('pdf')
```

```python
p1.name
```

```python
p1.show()
```

```python
w1=Word('word')
```

```python
w1.name
```

```python
w1.show()
```
